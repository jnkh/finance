```julia
using ZipFile, Requests, DataFrames
```

```julia
year = 2014
month = 12
month_str = "/12%20December"

destination_folder = "../data"
year_str = "/" * string(year)

dest = destination_folder * year_str * "/" * string(month)


curr1 = "EUR"
curr2 = "USD"
week = 1

filename = curr1 * "_" * curr2 * "_" * "Week$week"

zipped_ending = ".zip"
csv_ending = ".csv"


src_url = "ratedata.gaincapital.com"
```

```julia
zipped = get("http://" * src_url * year_str * month_str * "/" * filename *zipped_ending)
save(zipped,dest * "/" * filename * zipped_ending)
run(`unzip $(dest * "/" * filename * zipped_ending) -d $dest`)
```

```julia
df = readtable(dest * "/" * filename * csv_ending)
```

```julia
df.columns
```

```julia
function as_date_time(s::AbstractString)
    DateTime(s,"yyyy-mm-dd HH:MM:SS.sssssssss")
end

@vectorize_1arg AbstractString as_date_time

function convert_to_logspace(vec)
    ret_vec = similar(vec)
    ret_vec[1] = 0.0
    for i = 1:length(vec)-1
        ret_vec[i+1] = log(vec[i+1]/vec[i])
    end
    ret_vec
end
```

```julia
df[:,4] = as_date_time(df[:,4])
```

```julia
len = 10000
plot(df[1:len,4],cumsum(convert_to_logspace(df[1:len,6])))
plot(df[1:len,4],cumsum(convert_to_logspace(df[1:len,5])))
```

```julia
PyPlot.plt[:hist](convert_to_logspace(df[1:end,5]),bins=200,normed=false,log=true,alpha=0.4)
PyPlot.plt[:hist](convert_to_logspace(df[1:end,6]),bins=200,normed=false,log=true,alpha=0.4)

axvline(mean(convert_to_logspace(df[1:end,5])),color="r")
ylim([0.1,1e4])
xlim([-0.001,+0.001])
xlabel("Log change")
ylabel("Frequency")
```
